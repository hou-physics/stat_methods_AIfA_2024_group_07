## Problem 1: Fitting stellar luminosities

Let us assume that we measure the luminosity as a function of time for four samples of stars. Each sample contains 100 stars and we measure their luminosity at 50 times each. The recorded values for the stellar luminosities for four samples are in the `data` directory and designated `starcat_1.npy`, `starcat_2.npy`, `starcat_3.npy` and `starcat_4.npy`. The values are given in units of $L_\odot$.

Our hypothesis is that all stars in these four catalogs are non-variable, i.e. their luminosity does not vary as a function of time, and that the observed luminosities are normally distributed with a standard deviation of $\sigma(L_*)=0.1L_*$.

(i) Test if the data is consistent with this hypothesis by comparing the data to model predictions and computing the $\chi^2$ values of the observed data with respect to the assumed model.

(ii) Try to diagnose the reasons for the discrepancies found (if any).

In the next three problems, we want to analyze seven astrophysical catalogs containing the magnitudes of galaxies in three different wavelength bands. The goal is to find out if some of the datasets are not consistent with each other.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

data1 = np.load('data/starcat_1.npy')
data2 = np.load('data/starcat_2.npy')
data3 = np.load('data/starcat_3.npy')
data4 = np.load('data/starcat_4.npy')

score1 = 0
score2 = 0
score3 = 0
score4 = 0


def calculate_bin_boundaries(data, num_bins=10):
    min_val, max_val = np.min(data), np.max(data)
    bin_edges = np.linspace(min_val, max_val, num_bins + 1)
    return bin_edges

def calculate_bin_probabilities(bin_edges, mean, std_dev):
    bin_probs = []
    for i in range(len(bin_edges) - 1):
        # Calculate the cumulative probability at the bin edges
        p_left = stats.norm.cdf(bin_edges[i], mean, std_dev)
        p_right = stats.norm.cdf(bin_edges[i + 1], mean, std_dev)
        bin_prob = p_right - p_left
        bin_probs.append(bin_prob)
    return np.array(bin_probs)


def chi_square(data):
    mean = np.mean(data)
    bins = calculate_bin_boundaries(data, 10)
    bin_data = np.histogram(data, 10)[0]
    prob = calculate_bin_probabilities(bins, mean, 0.1 * mean)
    expect = prob * 50
    chi_square = stats.chisquare(bin_data, np.sum(bin_data)/np.sum(expect) * expect)
    return chi_square


for i in range(0, 50):
    test = chi_square(data1[i,:])[1]
    if test < 0.05:
        score1 = score1 + 1
        
for i in range(0, 50):
    test = chi_square(data2[i,:])[1]
    if test < 0.05:
        score2 = score2 + 1

for i in range(0, 50):
    test = chi_square(data3[i,:])[1]
    if test < 0.05:
        scor3 = score3 + 1
        
for i in range(0, 50):
    test = chi_square(data4[i,:])[1]
    if test < 0.05:
        score4 = score4 + 1

print(score1, score2, score3, score4) # high score means bad model

```

With 95% confindence, the hypothesis cannot be rejected for 44 stars in the first catalog and for all stars in the third one. On the other hand, the hypothesis works really poorly for catalog 2 and 4, where the hypothesis does not hold in 94 out of 100 cases. 

## Problem 2: Magnitude distributions and colors

Use `astropy.io.fits` to load the six fits files called `magnitudes0.fits` to `magnitudes5.fits`. Find out how to extract the magnitudes of the galaxies in each of the files in $g$, $r$, and $i$ bands. Write a function that takes the fits object and returns the colors:
$$C_1 =  \mathrm{mag}_g - \mathrm{mag}_r$$
$$C_2 =  \mathrm{mag}_i - \mathrm{mag}_r$$
Plot histograms of the colors and the color-color plane of each of the catalogs. 

Can you spot the non-consistent catalogs by eye?

```python
from astropy.io import fits

file_name = "data/magnitudes0.fits"  

#check the structure of fits data
with fits.open(file_name) as hdul:
    
    hdul.info()  
   
    data = hdul[1].data  
    print("Available columns:", data.columns.names)  
```

```python
import numpy as np
import matplotlib.pyplot as plt

# List of FITS files (update the paths if needed)
fits_files = [f"data/magnitudes{i}.fits" for i in range(6)]

# Function to calculate colors C1 and C2
def calculate_colors(fits_file):
    with fits.open(fits_file) as hdul:
        data = hdul[1].data  
        mag_g = data['G MAG']  # Magnitude in g-band
        mag_r = data['R MAG']  # Magnitude in r-band
        mag_i = data['I MAG']  # Magnitude in i-band

        # Calculate colors
        C1 = mag_g - mag_r
        C2 = mag_i - mag_r

        return C1, C2

# Initialize storage for colors
colors_C1 = []
colors_C2 = []

# Calculate colors for each FITS file
for file in fits_files:
    try:
        C1, C2 = calculate_colors(file)
        colors_C1.append(C1)
        colors_C2.append(C2)
    except Exception as e:
        print(f"Error processing {file}: {e}")

# Plot histograms for each catalog
for i, (C1, C2) in enumerate(zip(colors_C1, colors_C2)):
    plt.figure(figsize=(12, 6))
    
    # Histogram of C1
    plt.subplot(1, 2, 1)
    plt.hist(C1, bins=50, alpha=0.7, label=f'Catalog {i}')
    plt.xlabel('C1 (g-r)')
    plt.ylabel('Frequency')
    plt.legend()

    # Histogram of C2
    plt.subplot(1, 2, 2)
    plt.hist(C2, bins=50, alpha=0.7, label=f'Catalog {i}')
    plt.xlabel('C2 (i-r)')
    plt.ylabel('Frequency')
    plt.legend()
    
    plt.show()

# Plot color-color planes
for i, (C1, C2) in enumerate(zip(colors_C1, colors_C2)):
    plt.figure(figsize=(8, 8))
    hb = plt.hexbin(C1, C2, gridsize=50, cmap='viridis', mincnt=1)
    plt.colorbar(hb, label='Density')  # Add colorbar
    
    plt.xlabel('C1 (g-r)')
    plt.ylabel('C2 (i-r)')
    plt.title(f'Color-Color Plane for Catalog {i}')
    plt.legend()
    plt.show()


```

Maybe Catalog 5 is inconsistent because the histogram of C2 does not show a sudden sharp peak but rather a gradual increase. And from the c-c-plane,it also doesn't show a clear correlation between C1 and C2.

## Problem 3: T-test

Use the t-test (from `scipy.stats.ttest_ind`, for example) to compare both $C_1$ and $C_2$ between the different catalogs. 

Which catalogs are non-consistent with 95% confidence? 

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from itertools import combinations

nums = [0,1,2,3,4,5]
for i, j in combinations(nums, 2):
    ttest_C1 = stats.ttest_ind(colors_C1[i], colors_C1[j])
    ttest_C2 = stats.ttest_ind(colors_C2[i], colors_C2[j])
    if ttest_C1[1] < 0.05:
        print(f'Catalog {i} is inconsistent with catalog {j} based on C1')
    if ttest_C2[1] < 0.05:
        print(f'Catalog {i} is inconsistent with catalog {j} based on C2')
```

Based on the T-test, catalog 3 is inconsistent with the rest when it comes to C2.

## Problem 4: KS-test

Use the KS-test (from `scipy.stats.ks_2samp`, for example) to compare both $C_1$ and $C_2$ between the different catalogs. 

(i) Which catalogs are non-consistent with 95% confidence? 

(ii) How do you explain the discrepancies between findings of problem 3 and 4?

```python
from scipy.stats import ks_2samp

# Store results for non-consistent catalogs
non_consistent_catalogs_C1_ks = []
non_consistent_catalogs_C2_ks = []

# Perform pairwise KS tests for C1 and C2
for i in range(len(colors_C1)):
    for j in range(i + 1, len(colors_C1)):  # Only compare each pair once
        # KS test for C1
        ks_stat_C1, p_value_C1 = ks_2samp(colors_C1[i], colors_C1[j])
        if p_value_C1 < 0.05:
            non_consistent_catalogs_C1_ks.append((i, j))
        
        # KS test for C2
        ks_stat_C2, p_value_C2 = ks_2samp(colors_C2[i], colors_C2[j])
        if p_value_C2 < 0.05:
            non_consistent_catalogs_C2_ks.append((i, j))

# Print results
print("Non-consistent catalogs based on C1 (g-r) [KS-test]:", non_consistent_catalogs_C1_ks)
print("Non-consistent catalogs based on C2 (i-r) [KS-test]:", non_consistent_catalogs_C2_ks)
```

Based on the results of the KS-test, we found that Cat 5 is non-consistent not only with C2 but also with C1 when compared to other catalogs, which aligns with the initial assumption by eye. In addition, Cat 3 also shows its non-consitence based on C2, which aligns with the results of the T-test. The reason of the difference between those 2 test is the following:
The T-test checks for significant differences in averages, assuming normal distribution, but may miss shape or tail differences. The KS-test detects overall distribution differences, including shape and tails, without assuming any specific distribution.
