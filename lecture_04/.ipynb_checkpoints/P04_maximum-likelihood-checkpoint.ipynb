# P04: Maximum likelihood methods

In this challenge, we are going to study the determination of photometric redshifts in galaxy surveys. 

Galaxy redshifts are usually determined by comparing the observed galaxy spectrum to templates and inferring the redshift from the observed wavelength shift of the observed spectrum. Redshifts determined in this way are called *spectroscopic redshifts*. 

Since taking galaxy spectra is time-consuming, large surveys usually cannot afford to measure the full galaxy spectrum, but only the observed flux in discrete wavelength bands. Such photometric galaxy redshifts are then inferred by comparing the observed integrated flux in these bands to that expected for different templates. 

Photometric redshifts are more noisy (and more prone to biases) than spectroscopic ones, but since they require less observing time, photometric surveys are able to observe much larger galaxy numbers. This is particularly important for surveys focused on weak gravitational lensing, as for example the recently completed [Dark Energy Survey (DES)](https://www.darkenergysurvey.org/). The photometric bands for DES cover the visible and a bit of the infrared spectrum and are shown in the figure below.

## Problem 1: Finding the best-fit template spectrum amplitude

Let us assume that in restframe, all galaxies have a spectrum as the one given in `tspec[0]` but they vary in their total luminosity. Observationally we measure galaxy magnitudes, i.e. the logarithm of observed galaxy fluxes, in the five photometric bands given above. 

In this first problem, we will assume that we do not have a model for the measurement errors on galaxy magnitudes. As a first guess, we can therefore assume that the magnitudes (i.e. the log of the flux) are Gaussian-distributed with equal errors in each photometric band. Allowing for a variable galaxy luminosity in this model amounts to introducing a constant amplitude $A$ that consistently scales the model log-fluxes in all photometric bands.

(i) Write down an analytic expression for the likelihood that needs to be minimzed to find the best-fit amplitude $A$ for a given galaxy.

(ii) Assume that a galaxy at redshift $z=0$ yields the measured fluxes in our five photometric bands given in `galaxy_fluxes_z=0.txt`. Use the expression of the likelihood you have derived to determine the amplitude of the template log-fluxes that best fits the observations. 

### (i)
$m_{\text{obs}, i}$: Observed magnitude in photometric band $i$.

$m_{\text{model}, i}(A)$: Model magnitude in photometric band $i$ as a function of amplitude $A$.

$F_{\text{model}, i}$: Model flux in band $i$, where $F_{\text{model}, i} = A \times F_{0, i}$ and $F_{0, i}$ is the template flux.


The magnitude is related to flux by:
   $$
   m = -2.5 \log_{10} F + \text{const}
   $$
   
The model magnitude can be expressed as:
   $$
   m_{\text{model}, i}(A) = -2.5 \log_{10} (A F_{0, i}) + \text{const} = m_{0, i} - 2.5 \log_{10} A
   $$
where $m_{0, i} = -2.5 \log_{10} F_{0, i} + \text{const}$.

The difference between observed and model magnitudes is:
   $$
   m_{\text{obs}, i} - m_{\text{model}, i}(A) = (m_{\text{obs}, i} - m_{0, i}) + 2.5 \log_{10} A
   $$
   
   
Assuming Gaussian errors with equal variance $\sigma^2$, the negative log-likelihood function (up to a constant) is:
   $$
   \chi^2(A) = \sum_{i} \left( m_{\text{obs}, i} - m_{\text{model}, i}(A) \right)^2
   $$
Substitute the expression for the magnitude difference:
   $$
   \chi^2(A) = \sum_{i} \left( [m_{\text{obs}, i} - m_{0, i}] + 2.5 \log_{10} A \right)^2
   $$
   
An explicit expression for the likelihood is:
$$
\chi^2(A) = \sum_{i} \left( [m_{\text{obs}, i} - m_{0, i}] + 2.5 \log_{10} A \right)^2
$$
which we minimize over $A$.

```python
import numpy as np
from astropy.io import fits

# Load the template spectrum from the FITS file
path = 'data/k_nmf_derived.newdefault.fits'
temps = fits.open(path)
tspec = temps[1].data
template = tspec[0]  # Using the first template spectrum

# Resample the template to match the number of photometric bands (assuming 5 bands)
template_resampled = np.interp(np.linspace(0, len(template)-1, 5), np.arange(len(template)), template)

# Load the observed fluxes for the galaxy at z=0 from the provided file
fluxes_file_path = 'data/galaxy_fluxes_z=0.txt'
observed_fluxes = np.loadtxt(fluxes_file_path)

# Convert observed and template fluxes to magnitudes
m_obs = -2.5 * np.log10(observed_fluxes)
m_0 = -2.5 * np.log10(template_resampled)

# Define a range of amplitudes to test
amplitude_range = np.linspace(0.1, 10, 1000)  # Adjust the range as needed

# Calculate chi-squared for each amplitude
chi_squared_values = []
for A in amplitude_range:
    m_model = m_0 - 2.5 * np.log10(A)
    chi_squared = np.sum((m_obs - m_model)**2)
    chi_squared_values.append(chi_squared)

# Find the amplitude that minimizes chi-squared
best_amplitude = amplitude_range[np.argmin(chi_squared_values)]

# Display the best-fit amplitude
best_amplitude
```

## Problem 2: Photometric redshifts without error model

We now extend our model from problem 1 to determine galaxy redshifts: we can estimate a galaxy's redshift by solving the maximum likelihood problem from our first exercise on a grid of redshifts. The redshift which corresponds to the solution with the smallest residuals is our maximum-likelihood redshift estimate for the galaxy.

(i) Estimate the redshift of all galaxies in catalog `catalog1.txt`, assuming that all galaxies are of type `tspec[0]` and that their redshifts lie between $0 \leq z \leq 1.5$. 

(ii) Verify your results with the true redshifts in `redshifts1.txt`.

```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt

# Load the template spectrum from the FITS file
path = 'data/k_nmf_derived.newdefault.fits'
temps = fits.open(path)
tspec = temps[1].data
template = tspec[0]  # Using the first template spectrum

# Resample the template to match the number of photometric bands (assuming 5 bands)
template_resampled = np.interp(np.linspace(0, len(template)-1, 5), np.arange(len(template)), template)

# Load observed fluxes for all galaxies from catalog1.txt
catalog_file_path = 'data/catalog1.txt'
observed_fluxes_all = np.loadtxt(catalog_file_path)

# Convert template fluxes to magnitudes
m_0 = -2.5 * np.log10(template_resampled)

# Create a grid of redshift values between 0 and 1.5
redshift_grid = np.linspace(0, 1.5, 151)

# Function to calculate chi-squared for a given redshift with wavelength shifting
def chi_squared_z(z, observed_fluxes, m_0):
    # Shift the template for the redshift (simulate wavelength shift)
    template_shifted = template_resampled / (1 + z)
    m_model_z = -2.5 * np.log10(template_shifted)

    # Convert observed fluxes to magnitudes
    m_obs = -2.5 * np.log10(observed_fluxes)

    # Avoid invalid or infinite values
    valid_indices = np.isfinite(m_obs) & np.isfinite(m_model_z)
    m_obs = m_obs[valid_indices]
    m_model_z = m_model_z[valid_indices]

    return np.sum((m_obs - m_model_z)**2)

# Iterate over each galaxy's observed fluxes and find the best-fit redshift
best_redshifts = []
for observed_fluxes in observed_fluxes_all:
    chi_squared_values = []
    for z in redshift_grid:
        chi_squared = chi_squared_z(z, observed_fluxes, m_0)
        chi_squared_values.append(chi_squared)
    
    # Find the redshift that gives the minimum chi-squared value
    best_redshift = redshift_grid[np.argmin(chi_squared_values)]
    best_redshifts.append(best_redshift)

# Load the true redshifts from redshifts1.txt
true_redshifts_path = 'data/redshifts1.txt'
true_redshifts = np.loadtxt(true_redshifts_path)

# Plot the estimated redshifts vs true redshifts
plt.figure(figsize=(10, 6))
plt.scatter(true_redshifts, best_redshifts, color='blue', label='Estimated vs True Redshifts')
plt.plot([0, 1.5], [0, 1.5], 'r--', label='Ideal Line')
plt.xlabel('True Redshift')
plt.ylabel('Estimated Redshift')
plt.title('Comparison of Estimated Redshifts with True Redshifts')
plt.legend()
plt.grid(True)
plt.show()
```

## Problem 3: Photometric redshifts with error model

Let us now assume that we have a model for our errors, i.e. we can find a full expression for our likelihood.

(i) Estimate the redshifts of the galaxies in catalogs `catalog1.txt`, `catalog2.txt`, and `catalog3.txt` assuming a Gaussian error on the log of the flux in each band of $(\sigma_g, \sigma_r, \sigma_i, \sigma_z, \sigma_Y) = (0.15, 0.1, 0.08, 0.05, 0.03)$ for catalog 1, $(\sigma_g, \sigma_r, \sigma_i, \sigma_z, \sigma_Y) = (1.0, 0.8, 0.6, 0.4, 0.2)$ for catalog 2 and $(\sigma_g, \sigma_r, \sigma_i, \sigma_z, \sigma_Y) = (0.5, 0.4, 0.3, 0.2, 0.1)$ for catalog 3. 

(ii) Analyze your results by plotting data and best-fit model with errorbars for a random selection of galaxies from the catalog. What do you conclude about the error model for catalogs `catalog1.txt`, `catalog2.txt`, and `catalog3.txt`?

```python
import numpy as np
from astropy.io import fits
import matplotlib.pyplot as plt

# Load the template spectrum from the FITS file
path = 'data/k_nmf_derived.newdefault.fits'
temps = fits.open(path)
tspec = temps[1].data
template = tspec[0]  # Using the first template spectrum

# Resample the template to match the number of photometric bands (assuming 5 bands)
template_resampled = np.interp(np.linspace(0, len(template)-1, 5), np.arange(len(template)), template)

# Load observed fluxes for all galaxies from the three catalogs
catalog_files = ['data/catalog1.txt',
                 'data/catalog2.txt',
                 'data/catalog3.txt']

# Define Gaussian errors for each catalog
errors_catalogs = {
    'catalog1.txt': np.array([0.15, 0.1, 0.08, 0.05, 0.03]),
    'catalog2.txt': np.array([1.0, 0.8, 0.6, 0.4, 0.2]),
    'catalog3.txt': np.array([0.5, 0.4, 0.3, 0.2, 0.1])
}

# Convert template fluxes to magnitudes
m_0 = -2.5 * np.log10(template_resampled)

# Create a grid of redshift values between 0 and 1.5
redshift_grid = np.linspace(0, 1.5, 151)

# Function to calculate chi-squared for a given redshift with error model
def chi_squared_z_with_errors(z, observed_fluxes, m_0, errors):
    # Shift the template for the redshift (simulate wavelength shift)
    template_shifted = template_resampled / (1 + z)
    m_model_z = -2.5 * np.log10(template_shifted)

    # Convert observed fluxes to magnitudes
    m_obs = -2.5 * np.log10(observed_fluxes)

    # Avoid invalid or infinite values
    valid_indices = np.isfinite(m_obs) & np.isfinite(m_model_z)
    m_obs = m_obs[valid_indices]
    m_model_z = m_model_z[valid_indices]
    errors = errors[valid_indices]

    return np.sum(((m_obs - m_model_z) / errors)**2)

# Iterate over each catalog
results = {}
for catalog_file in catalog_files:
    catalog_name = catalog_file.split('/')[-1]
    observed_fluxes_all = np.loadtxt(catalog_file)
    errors = errors_catalogs[catalog_name]

    best_redshifts = []
    for observed_fluxes in observed_fluxes_all:
        chi_squared_values = []
        for z in redshift_grid:
            chi_squared = chi_squared_z_with_errors(z, observed_fluxes, m_0, errors)
            chi_squared_values.append(chi_squared)
        
        # Find the redshift that gives the minimum chi-squared value
        best_redshift = redshift_grid[np.argmin(chi_squared_values)]
        best_redshifts.append(best_redshift)
    
    results[catalog_name] = best_redshifts


# (ii) Analyze and plot results for a random selection of galaxies
num_samples = 5  # Number of random samples to plot
for catalog_file in catalog_files:
    catalog_name = catalog_file.split('/')[-1]
    observed_fluxes_all = np.loadtxt(catalog_file)
    errors = errors_catalogs[catalog_name]
    best_redshifts = results[catalog_name]

    # Plot random samples
    plt.figure(figsize=(12, 6))
    sample_indices = np.random.choice(len(observed_fluxes_all), num_samples, replace=False)
    for i in sample_indices:
        observed_fluxes = observed_fluxes_all[i]
        m_obs = -2.5 * np.log10(observed_fluxes)
        best_z = best_redshifts[i]
        m_model_best = m_0 - 2.5 * np.log10(1 + best_z)

        plt.errorbar(range(5), m_obs, yerr=errors, fmt='o', label=f'Galaxy {i+1} Observed', capsize=3)
        plt.plot(range(5), m_model_best, label=f'Galaxy {i+1} Model (z={best_z:.2f})')

    plt.title(f'Comparison of Observed Data and Best-Fit Model for {catalog_name}')
    plt.xlabel('Photometric Band')
    plt.ylabel('Magnitude')
    plt.legend()
    plt.show()
```
