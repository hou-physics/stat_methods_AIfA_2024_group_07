## Problem 1: Fisher matrix for a Gaussian likelihood

Assume data $\boldsymbol{x}$ with mean $\boldsymbol{\mu}$ and covariance $\boldsymbol{\mathrm{C}}$, following an $d$-dimensional Gaussian distribution in data space. Recall that $\langle\boldsymbol{x}\rangle =\boldsymbol{\mu}$ and $\langle\boldsymbol{\mathrm{D}}\rangle \coloneqq \langle(\boldsymbol{x} - \boldsymbol{\mu})\otimes (\boldsymbol{x} - \boldsymbol{\mu})\rangle = \boldsymbol{\mathrm{C}}$. 

1. Show that in this case, the Fisher matrix can be written as:
$$
\boldsymbol{\mathrm{F}}_{\alpha\beta} = \frac{1}{2}\mathrm{tr}\left(\boldsymbol{\mathrm{C}}^{-1} \frac{\partial\boldsymbol{\mathrm{C}}}{\partial\boldsymbol{\theta}^\alpha}\boldsymbol{\mathrm{C}}^{-1} \frac{\partial\boldsymbol{\mathrm{C}}}{\partial\boldsymbol{\theta}^\beta} + 2\boldsymbol{\mathrm{C}}^{-1} \frac{\partial \boldsymbol{\mu}}{\partial\boldsymbol{\theta}^\alpha} \frac{\partial \boldsymbol{\mu}^\mathrm{T}}{\partial\boldsymbol{\theta}^\beta} \right)\,. 
$$

2. Show that any linear transformation of the data will leave the Fisher matrix unchanged. Hint: any vector, $\boldsymbol{x}$, changes as $\boldsymbol{y} = \boldsymbol{\mathrm{A}}\boldsymbol{x}$ under a linear transformation, where $\boldsymbol{\mathrm{A}}$ is the transformation matrix.

## Solution 1:
### 1.1

We start by noting that since $\boldsymbol{\mathrm{C}}$ is symmetric, so are all its derivatives. Furthermore, we will use the 
$$
-\ln L = \frac{1}{2}\mathrm{tr}\left(\ln \boldsymbol{\mathrm{C}} + \boldsymbol{\mathrm{C}}^{-1}\boldsymbol{\mathrm{D}}\right) + \mathrm{const}\,,
$$
from the lecture. Noticing that $\partial_\alpha(\boldsymbol{\mathrm{C}}^{-1}) = - \boldsymbol{\mathrm{C}}^{-1}\partial_\alpha \boldsymbol{\mathrm{C}} \boldsymbol{\mathrm{C}}^{-1}$ we write:
$$
- 2\partial_\alpha\ln L = \mathrm{tr}\left(\boldsymbol{\mathrm{C}}^{-1}\partial_\alpha \boldsymbol{\mathrm{C}} - \boldsymbol{\mathrm{C}}^{-1}(\partial_\alpha \boldsymbol{\mathrm{C}})\boldsymbol{\mathrm{C}}^{-1} \boldsymbol{\mathrm{D}} + \boldsymbol{\mathrm{C}}^{-1} \partial_\alpha \boldsymbol{\mathrm{D}} \right)\;.
$$
Taking the second derivative gives
$$
- 2\partial_\alpha\partial_\beta\ln L =  \mathrm{tr}\Big( -\boldsymbol{\mathrm{C}}^{-1}  \partial_\alpha \boldsymbol{\mathrm{C}} \boldsymbol{\mathrm{C}}^{-1}  \partial_\beta \boldsymbol{\mathrm{C}} + \boldsymbol{\mathrm{C}}^{-1}  \partial_\alpha\partial_\beta \boldsymbol{\mathrm{C}} + \boldsymbol{\mathrm{C}}^{-1}\left[ \partial_\alpha \boldsymbol{\mathrm{C}} \boldsymbol{\mathrm{C}}^{-1}\partial_\beta \boldsymbol{\mathrm{C}}  +  \partial_\beta \boldsymbol{\mathrm{C}} \boldsymbol{\mathrm{C}}^{-1}\partial_\alpha \boldsymbol{\mathrm{C}}  \right]\boldsymbol{\mathrm{C}}^{-1} \boldsymbol{\mathrm{D}} \\
 - \boldsymbol{\mathrm{C}}^{-1}\left[ \partial_\alpha \boldsymbol{\mathrm{C}} \boldsymbol{\mathrm{C}}^{-1} \partial_\beta \boldsymbol{\mathrm{D}} +  \partial_\beta\boldsymbol{\mathrm{C}} \boldsymbol{\mathrm{C}}^{-1} \partial_\alpha \boldsymbol{\mathrm{D}}   \right] -  \boldsymbol{\mathrm{C}}^{-1} \partial_\alpha \partial_\beta \boldsymbol{\mathrm{C}} \boldsymbol{\mathrm{C}}^{-1}\boldsymbol{\mathrm{D}} + \boldsymbol{\mathrm{C}}^{-1}\partial_\alpha \partial_\beta \boldsymbol{\mathrm{D}} \Big )

$$
At the true parameter values we have $\boldsymbol{\mathrm{C}} = \langle\boldsymbol{\mathrm{D}}\rangle$ and since $\langle \boldsymbol{x}\rangle = \boldsymbol{\mu}$, we have

$$
\langle\partial_\alpha\boldsymbol{\mathrm{D}}\rangle \sim - \partial_\alpha \boldsymbol{\mu}(\langle\boldsymbol{x}\rangle - \boldsymbol{\mu}) = 0 \\
\langle \partial_\alpha\partial_\beta\boldsymbol{\mathrm{D}}\rangle = \partial_\alpha \boldsymbol{\mu} \partial_\beta \boldsymbol{\mu}^\mathrm{T} +  \partial_\beta \boldsymbol{\mu} \partial_\alpha \boldsymbol{\mu}^\mathrm{T}\,.
$$

Using these expressions and taking the average of the equation of the second derivatives, gives the Fisher matrix:

$$
F_{\alpha\beta} = -\langle \partial_\alpha\partial_\beta\ln L \rangle = \frac{1}{2}\mathrm{tr}\Big(  -\boldsymbol{\mathrm{C}}^{-1}  \partial_\alpha \boldsymbol{\mathrm{C}} \boldsymbol{\mathrm{C}}^{-1}  \partial_\beta \boldsymbol{\mathrm{C}} + \boldsymbol{\mathrm{C}}^{-1}  \partial_\alpha\partial_\beta \boldsymbol{\mathrm{C}} + \boldsymbol{\mathrm{C}}^{-1}\left[ \partial_\alpha \boldsymbol{\mathrm{C}} \boldsymbol{\mathrm{C}}^{-1}\partial_\beta \boldsymbol{\mathrm{C}}  +  \partial_\beta \boldsymbol{\mathrm{C}} \boldsymbol{\mathrm{C}}^{-1}\partial_\alpha \boldsymbol{\mathrm{C}}  \right] \\
-  \boldsymbol{\mathrm{C}}^{-1} \partial_\alpha \partial_\beta \boldsymbol{\mathrm{C}}  + \boldsymbol{\mathrm{C}}^{-1}\left[\partial_\alpha \boldsymbol{\mu} \partial_\beta \boldsymbol{\mu}^\mathrm{T} +  \partial_\beta \boldsymbol{\mu} \partial_\alpha \boldsymbol{\mu}^\mathrm{T}\right] \Big)\,,
$$

cancelling additional terms:

$$
F_{\alpha\beta} = \frac{1}{2}\mathrm{tr}\Big( \boldsymbol{\mathrm{C}}^{-1} \partial_\beta \boldsymbol{\mathrm{C}} \boldsymbol{\mathrm{C}}^{-1}\partial_\alpha \boldsymbol{\mathrm{C}}  
 + \boldsymbol{\mathrm{C}}^{-1}\left[\partial_\alpha \boldsymbol{\mu} \partial_\beta \boldsymbol{\mu}^\mathrm{T} +  \partial_\beta \boldsymbol{\mu} \partial_\alpha \boldsymbol{\mu}^\mathrm{T}\right] \Big)\,.

$$

Since the second derivative is symmetric (also because the trace is cyclic), we can permute $\alpha$ and $\beta$. Furthermore we note that

$$
\mathrm{tr}\left(\boldsymbol{\mathrm{C}}^{-1}\left[\partial_\alpha \boldsymbol{\mu} \partial_\beta \boldsymbol{\mu}^\mathrm{T} +  \partial_\beta \boldsymbol{\mu} \partial_\alpha \boldsymbol{\mu}^\mathrm{T}\right] \right) = 2 \mathrm{tr}\left(\mathrm{C}^{-1} \partial_\alpha \boldsymbol{\mu} \partial_\beta \boldsymbol{\mu}^\mathrm{T}\right)\,,
$$
which yields the desired result.


### 1.2
Under such a linear transformation, the mean transforms as: $\boldsymbol{\mu}_y = \boldsymbol{\mathrm{A}} \boldsymbol{\mu}_x $ and the covariance as $\boldsymbol{\mathrm{C}}_y = \boldsymbol{\mathrm{A}}\boldsymbol{\mathrm{C}}_x \boldsymbol{\mathrm{A}}^\mathrm{T}$. It is also easy to see that:

$$
\boldsymbol{\mathrm{C}}^{-1}_y = \boldsymbol{\mathrm{A}}^{\mathrm{T}-1}\boldsymbol{\mathrm{C}}^{-1}_x \boldsymbol{\mathrm{A}}^{-1}\;,
$$

plugging everything into the expression of the Fisher matrix we see that, for example for the first term:

$$
\mathrm{tr}\left(\boldsymbol{\mathrm{A}}^{\mathrm{T}-1}\boldsymbol{\mathrm{C}}^{-1}_x \boldsymbol{\mathrm{A}}^{-1}\partial_\alpha \boldsymbol{\mathrm{A}}\boldsymbol{\mathrm{C}}_x \boldsymbol{\mathrm{A}}^\mathrm{T}\boldsymbol{\mathrm{A}}^{\mathrm{T}-1}\boldsymbol{\mathrm{C}}^{-1}_x \boldsymbol{\mathrm{A}}^{-1}\partial_\beta \boldsymbol{\mathrm{A}}\boldsymbol{\mathrm{C}}_x \boldsymbol{\mathrm{A}}^\mathrm{T} \right) = \mathrm{tr}\left(\boldsymbol{\mathrm{A}}^{\mathrm{T}-1}\boldsymbol{\mathrm{C}}^{-1}_x \partial_\alpha\boldsymbol{\mathrm{C}}_x \boldsymbol{\mathrm{C}}^{-1}_x\partial_\beta \boldsymbol{\mathrm{C}}_x \boldsymbol{\mathrm{A}}^\mathrm{T} \right)
=  \mathrm{tr}\left(\boldsymbol{\mathrm{C}}^{-1}_x \partial_\alpha\boldsymbol{\mathrm{C}}_x \boldsymbol{\mathrm{C}}^{-1}_x\partial_\beta \boldsymbol{\mathrm{C}}_x \right)\,,
$$
where we used the cyclic property of the trace in the last step. A similar argument can be made for the second term of the Fisher matrix. Therefore, it does not change under linear transformations of the data, the information is conserved.



## Problem 2: Fisher matrix of supernovae and pitfalls

The code snippet below provides you with the luminosity distance used in cosmology. This measurement (ignoring some additional complication) is used in cosmology by observing supernovae of type Ia to deduce the expansion history of the Universe and therefore its content.
The code calculates the luminosity distance for $N_\mathrm{sn}$ and parameters $\theta$. The variable $N_\mathrm{trapz}$ controls the number of function evaluations in the numerical integration of the luminosity distance, the latter is given by:

$$
d_\mathrm{L}(a) = \frac{1}{a|K|^{1/2}}f_{K}\left(-|K|^{1/2}\chi_\mathrm{H}\int_1^{a}\frac{\mathrm{d}a'}{a'^2E(a')}\right)\,,
$$

where $\chi_\mathrm{H} = c/H_0$ is the Hubble radius $c$ the speed of light and $H_0$ the expansion rate of the Universe today. $K$ measures the spatial curvature and is given by $K= \Omega_\mathrm{k}/\chi_H^2$. The function $f_{K}$ is

$$

f_K(x) = \left\{\begin{array}{cc}
\sin(x) & \mathrm{if\;} K > 0\,,\\
x & \mathrm{if\;} K = 0\,,\\
\mathrm{sinh}(x) & \mathrm{if\;} K < 0\;.
\end{array}\right.
$$

Lastly, $E(a)$ is the dimensionless Hubble function, also called expansion function. It is given by:

$$
E(a) = a^{-3/2}\left(\Omega_\mathrm{m} + \Omega_\Lambda a^{-3(w_0 + w_\mathrm{a})}\exp\left(3w_\mathrm{a}(a-1)\right) + \Omega_\mathrm{K}a\right)^{1/2}\,.
$$

The integral is carried out with a trapezoidal rule using $N_\mathrm{trapz}$ points.

The model consists of five parameters, which we bundle into the vector $\boldsymbol{\theta}$ such that:

$$
\boldsymbol{\theta}^\mathrm{T} = (\Omega_\mathrm{m}, \Omega_\mathrm{k},\chi_\mathrm{H}, w_0, w\mathrm{a})\,,\quad \boldsymbol{\theta}^\mathrm{T}_0 =   (0.3, 0., 4.0, -1,.0)\, 
$$

the meaning of this parameters is the fractional matter density in the Universe, the curvature density parameter, the Hubble radius (or the Hubble constant, as the speed of light is constant) and two parameters for dynamical dark energy.

Lastly note that I added a tiny random noise on the final value of the luminosity distance to mimic numerical noise.

1. Write a function to numerically calculate the derivatives of the luminosity distance with respect to all parameters. You can do this, using finite differencing: $$
\frac{\partial f(x)}{\partial x} = \frac{f(x + h) - f(x-h)}{2h}\,,$$ for some small $h$.

2. Assume that the data points are not correlated and have all a relative error of 10 per cent. Write a code to calculate the Fisher matrix by using the equation from problem 1 and ignore any parameter dependence in the covariance matrix. Matrix inversion can be done by `numpy.linalg.inv(matrix)`.

3. Plot the conditional and marginal uncertainties $1\sigma$ of all parameters as a function of the step width, $h$, used for the finite differencing. 

4. Show the 2d marginal $1\sigma$ uncertainty for the parameters $w_0$ and $w_\mathrm{a}$. 

```python
import numpy as np
import matplotlib.pyplot as plt

theta_0 = [0.3, 0., 4.0, -1,.0]



# luminosity distance as a function parameters
def luminosity_distance(theta, N_trapz, N_sn):


    np.random.seed(seed=42)
    # randomly chose redshifts between 0 and 1.5 for the SN
    z_sn = np.random.rand(N_sn)*1.5
    a_sn = 1/(1+z_sn)
    a_int = []

    np.random.seed()

    # integration variable
    for i_a, val_a in enumerate(a_sn):
        a_int.append(np.linspace(1,val_a, N_trapz))
    a_int = np.array(a_int)

    # expansion function as a function of parameters and scale factor
    Om = theta[0]
    Ok = theta[1]
    Ol = 1.0 - Ok - Om
    chih = theta[2]
    w0 = theta[3]
    wa = theta[4]
    eofa = a_int**-1.5*(Om + Ol*a_int**(-3*(w0 + wa))*np.exp(3*wa*(a_int-1)) + Ok*a_int)**.5
    chih = theta[2]
    Ok = theta[1]
    chi = -chih*np.trapezoid(1/(a_int**2*eofa), x = a_int) 
    chi += chi*np.random.rand(N_sn)*1e-6
    K = np.abs(Ok/chih**2)
    if Ok > 0:
        return 1/a_sn*K**(-.5)*np.sin(K**.5*chi)
    if Ok < 0:
        return 1/a_sn*K**(-.5)*np.sinh(K**.5*chi)
    if Ok == 0:
        return 1/a_sn*chi
```

## Solution 2.1

First, we calculate the derivative with finite differencing

```python
def get_derivatives(theta_0, h ,N_trapz, N_sn):    
    derivative = []
    for i_par, val_par in enumerate(theta_0):
        theta_p = np.copy(theta_0)
        theta_m = np.copy(theta_0)
        theta_p[i_par] += h
        theta_m[i_par] -= h
        derivative.append((luminosity_distance(theta_p, N_trapz, N_sn) - luminosity_distance(theta_m, N_trapz, N_sn))/2./h)
    derivative = np.array(derivative)
    return derivative
```

## Solution 2.2
Next, we use the derivative to calculate the Fisher matrix

```python

def get_fisher(h, N_trapz = int(1e3), theta_0 = [0.3, 0., 4.0, -1,.0], N_sn = 100):
    relative_error = 0.1 # ten percent relative error
    covariance = np.eye(N_sn)*relative_error**2*luminosity_distance(theta_0, N_trapz, N_sn)**2  # covariance matrix is diagonal
    derivative = get_derivatives(theta_0, h, N_trapz, N_sn)
    fisher = np.zeros((len(theta_0), len(theta_0)))
    for i_p in range(len(theta_0)):
        for j_p in range(len(theta_0)):
            fisher[i_p,j_p] = np.trace(derivative[i_p, :, None]*derivative[j_p, None, :]*np.linalg.inv(covariance))*2
    return fisher
```

## Solution 2.4

First we find the marginalised fisher matrix by removing all rows and columns, but the one for $w_0$ and $w_\mathrm{a}$ from the covariance, i.e. the inverse Fisher matrix.

```python
h_stable = 5e-3 # this should be stable according to the plot above
fish = get_fisher(h_stable)
cov_fish = np.linalg.inv(fish)
cov_fish_marg = cov_fish[3:,3:]
```

Next, we calculate the angle of the ellipse and the two major axis. If a 2 dimensional covariance is given by:

$$
\boldsymbol{\mathrm{C}}_\mathrm{2d}
=\begin{pmatrix}
\sigma^2_1 & \sigma^2_{12}\\
\sigma^2_{21} & \sigma^2_{2}
\end{pmatrix}\;,
$$

where $\sigma^2_{21} = \sigma^2_{12}$, these are given by

$$
a/b^2 = \frac{\sigma^2_1 + \sigma^2_1}{2}\pm \left(\frac{(\sigma^2_1 - \sigma^2_2)^2}{4} + \sigma^4_{12}   \right)^{1/2}
$$

and the angle:

$$
\tan(2\theta) = \frac{2\sigma^2_{12}}{\sigma^2_1-\sigma^2_2}
$$ 

```python

theta = -0.5*np.arctan(2*cov_fish_marg[0,1]/(cov_fish_marg[0,0] - cov_fish_marg[1,1])) # the minus is because we measure against w_a in the plot but calculate for w_0 here

p = cov_fish_marg[0,0] + cov_fish_marg[1,1]
q = cov_fish_marg[0,0] - cov_fish_marg[1,1]

a = (p/2 + np.sqrt(q**2/4 + cov_fish_marg[0,1]**2))**.5
b = (p/2 - np.sqrt(q**2/4 + cov_fish_marg[0,1]**2))**.5

from matplotlib.patches import Ellipse

factor = 1.52
fig, ax = plt.subplots()


el = Ellipse((0, -1), width=factor*2*a, height=factor*2*b, angle=  theta*180/np.pi) 
ax.add_artist(el)


ax.set_xlim(-2*a,2*a)
ax.set_ylim(-1-4*b,-1+4*b)
ax.set_xlabel(r"$w_\mathrm{a}$", fontsize = fontsi)
ax.set_ylabel(r"$w_0$", fontsize = fontsi)

plt.tight_layout()
plt.savefig("ellipse.pdf")

```
